# Анализ рынка подержанных автомобилей (craigslist)

Этот проект посвящен исследованию данных о продаже автомобилей. 
Мы используем датасет объемом **426 880 строк**, загружаемый из локального хранилища.

### Основные этапы:
1. **Загрузка данных**: Чтение файла `vehicles.csv` чанками по 100 000 строк.
2. **Очистка**: Удаление аномальных цен (выбросов) и пустых значений.
3. **Визуализация**: Построение графиков распределения цен и популярности марок.
4. **Анализ**: Сравнение характеристик разных моделей.

> **Заметка:** Путь к данным берется из файла `.env` через модуль `src.config`.

```python
import pandas as pd
from src.config import VEHICLES_PATH
from src.logger_config import setup_logging

logger = setup_logging()

df = pd.read_csv(VEHICLES_PATH)
logger.info(f"Dataset loaded. Shape: {df.shape}")
```

```python
df_sample = df.sample(frac=0.05, random_state=42)
logger.info(f"Размер выборки для анализа: {df_sample.shape}")
```

```python
mean_p = df_sample['price'].mean()
median_p = df_sample['price'].median()

logger.info(f"Средняя цена (Mean): {mean_p:.2f}$")
logger.info(f"Медианная цена (Median): {median_p:.2f}$")
logger.info(f"Разница: {mean_p - median_p:.2f}$")
```
