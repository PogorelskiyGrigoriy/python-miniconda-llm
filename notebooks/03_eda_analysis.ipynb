```python
import pandas as pd
from src.logger_config import setup_logging
from src.config import FINAL_DATA_PATH

logger = setup_logging()

df = pd.read_csv(FINAL_DATA_PATH)
```

```python
top_10_manufacturers = df['manufacturer'].value_counts().head(10)

logger.info("ТОП-10 производителей по количеству объявлений:")
logger.info(top_10_manufacturers)
```

```python
top_10_types = df['type'].value_counts().head(10)

logger.info(top_10_types)
```

```python
mean_price = df['price'].mean()
median_price = df['price'].median()

logger.info(f"mean price: {mean_price:.2f}")
logger.info(f"median price: {median_price:.2f}")
```

```python
mean_odometer = df['odometer'].mean()
median_odometer = df['odometer'].median()

logger.info(f"mean odometer: {mean_odometer:.2f}")
logger.info(f"median odometer: {median_odometer:.2f}")
```

```python
# 1. Находим самого популярного производителя
most_sold_manufacturer = df['manufacturer'].value_counts().idxmax()

# 2. Считаем количество проданных авто для него
vehicles_count = df['manufacturer'].value_counts().max()

# 3. Фильтруем данные только по этому производителю для расчета цен
most_sold_df = df[df['manufacturer'] == most_sold_manufacturer]

min_price = most_sold_df['price'].min()
max_price = most_sold_df['price'].max()

logger.info(f"Leader: {most_sold_manufacturer}, count: {vehicles_count}, price range: ${min_price}-${max_price}")
```
